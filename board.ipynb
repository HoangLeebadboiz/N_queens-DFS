```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, x):
        self.stack.append(x)

    def pop(self):
        if len(self.stack) == 0:
            return "Stack is empty"
        else:
            return self.stack.pop()
        
    def is_empty(self):
        return len(self.stack) == 0
    
    def len_of_stack(self):
        return len(self.stack)
    
    def call_stack(self):
        print("Stack include:", str(self.stack))
```
