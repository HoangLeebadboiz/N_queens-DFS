```python
class Stack:
    def __init__(self):
        self.stack = []

    def push(self, x):
        self.stack.append(x)

    def pop(self):
        if len(self.stack) == 0:
            return "Stack is empty"
        else:
            return self.stack.pop()
        
    def is_empty(self):
        return len(self.stack) == 0
    
    def len_of_stack(self):
        return len(self.stack)
    
    def call_stack(self):
        print("Stack include:", str(self.stack))
```

```python
class DFS:
    def __init__(self, n):
        self.N = n
        self.node = []
        self.Board = np.zeros((n, n), dtype= int)
        #self.Back_Board = np.zeros((n, n), dtype= int)
        self.queen = 0
        self.count = 0

    def check(self, position : list):
        # Khởi tạo thông số
        x = position[0]
        y = position[1]
        #queen = 0
        # Xây dựng mô hình
        for i in range(x, self.N):
            for j in range(self.N):
                self.count += 1
                if (i == x)and (j < y):
                    continue
                if len(self.node) == self.N:
                    return True
                elif self.Board[i, j] != 0:
                    continue
                else:
                    self.node.append([i, j])
                    self.queen += 1
                    self.Board[i, j] = self.queen
                    self.loaitru([i, j], self.queen)
                    y = j
                    #print(self.node)
                    #print(self.queen)

        if self.queen < self.N:
            #print(self.node)
            
            for i in range(self.N):
                for j in range(self.N):
                    if self.Board[i, j] == self.queen:
                        self.Board[i, j] = -(self.queen - 1)

                    if self.Board[i, j] == -self.queen:
                        self.Board[i, j] = 0
            self.queen -= 1
            x = self.node[len(self.node)-1][0]
            y = self.node[len(self.node)-1][1]
            self.node.pop(-1)
            #print(self.node)
            #print(self.Board)
            #print("===============================================")
            
            if y == (self.N - 1):
                self.check([x+1, 0])
            if x  == self.N:
                return False
                
            else:
                self.check([x, y+1])
        else:
            return True
            
                


    def loaitru(self, position : list, queen : int):
        x = position[0]
        y = position[1]
        #self.Back_Board = self.Board.copy()
        for i in range(self.N):
            for j in range(self.N):
                if(self.Board[i, j] != 0):
                    continue
                elif(i == x and j != y) or (j == y and i != x) or (i + j == x + y) or (i - j == x - y):
                    self.Board[i, j] = -self.queen
                else:
                    continue

        #print(self.Board)
        
```
